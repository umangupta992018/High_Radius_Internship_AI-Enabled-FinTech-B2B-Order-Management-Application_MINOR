```python
#QUESTION-2(a).Get the items not common to both series A and series B (Without using loops)??
#QUESTION-2(b).Get the items common to both series A and series B(Without using loops)??

# input
#ser1 = pd.Series([1, 2, 3, 4, 5])
#ser2 = pd.Series([4, 5, 6, 7, 8])??

# ANSWER OF QUESTION-2(a),(b)

ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
union=pd.Series(np.union1d(ser1,ser2))
intersection=pd.Series(np.intersect1d(ser1,ser2))
not_commom=union[~union.isin(intersection)]
print("Not common numbers are:")
print(not_commom)
print()
print("Commom numbers are: ")
print(intersection)
```

```python
#PART-B
#QUESTION-1. Find the Average price of project from each state??

# ANSWER OF QUESTION-1



myFile.groupby('school_state',sort=False)['price'].mean()
```

```python
#PART-B
#QUESTION-2.a. Find the total number of projects previously posted by all the teachers belonging to each teacher prefix.
#For Example all the teachers having prefix as dr have posted a total of 13 projects combined previously.??

# ANSWER OF QUESTION-2.a

myFile.groupby('teacher_prefix')['teacher_number_of_previously_posted_projects'].sum()
```

```python
#PART-B
#QUESTION-2.b. Find the prefix of the teacher who has posted the maximum of projects previously??

# ANSWER OF QUESTION-2.b

myFile.loc[myFile.teacher_number_of_previously_posted_projects.max(),'teacher_prefix']
```

```python
#PART-B
#QUESTION 3. Find the number of projects approved for each project subject category belonging to the project grade category 'grade_9_12'??

# ANSWER OF QUESTION-3

myFile1 = myFile.loc[myFile.project_grade_category == 'grades_9_12'].loc[myFile.project_is_approved == 1]
myFile1.groupby('project_subject_categories')['project_is_approved'].count().reset_index()
```

```python
#PART-B
#QUESTION 4.Replace teacher_prefix with the average number of approved projects for each teacher prefix??

# ANSWER OF QUESTION-4

myFile['teacher_prefix'] = myFile1.groupby('teacher_prefix')['project_is_approved'].transform('mean')
myFile
```

```python
#QUESTION 5.5.Find the desired output as shown in the image below.??
#Note : Create the data frame (Train Data) as shown in the image below??


# ANSWER OF QUESTION-5

Train_Data = pd.DataFrame({
    'State':['A', 'B', 'C', 'A', 'A', 'B', 'A', 'A', 'C', 'C'],
    'Class':[0, 1, 1, 0, 1, 1, 0, 1, 1, 0]
})
Train_Data
```

```python
#PART-B
#QUESTION 5.5.Find the desired output as shown in the image below.??
#Note : Create the data frame (Train Data) as shown in the image below??


# ANSWER OF QUESTION-5

response_table = Train_Data.groupby(['State', 'Class'])['Class'].count().unstack(level=-1).fillna(0)
response_table.columns=['class=0', 'class=1']
response_table.reset_index()
```

```python
#PART-B
#QUESTION 5.5.Find the desired output as shown in the image below.??
#Note : Create the data frame (Train Data) as shown in the image below??


# ANSWER OF QUESTION-5

class0 = response_table['class=0'] / response_table.sum(axis=1)
class1 = response_table['class=1'] / response_table.sum(axis=1)
state_0 = Train_Data['State'].apply(lambda x: class0[x])
state_1 = Train_Data['State'].apply(lambda x: class1[x])
data = {'State_0': state_0, 'State_1': state_1}
Encoded_Train_data = pd.DataFrame(data)
Encoded_Train_data
```
